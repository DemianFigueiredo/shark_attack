```python
import re
import numpy as np

# Assuming your DataFrame is named 'df' and the column you want to clean is named 'Year'

# Remove non-digit characters and keep only numbers
Data['Age'] = Data['Age'].astype(str).str.replace(r'\D+', '', regex=True)

# Replace occurrences of two numbers with their mean value
def replace_with_mean(value):
    numbers = re.findall(r'\d+', value)
    if len(numbers) == 2:
        mean = np.mean([int(num) for num in numbers])
        value = re.sub(r'\d+', str(mean), value)
    return value

Data['Age'] = Data['Age'].apply(replace_with_mean)

columns_to_check2 = ['Age']

# Iterate over the specified columns and print unique values
for column in columns_to_check2:
    unique_values2 = Data[column].unique()
    print("Unique values of column '{}': {}".format(column, unique_values2))
```

```python
import re
import numpy as np

# Assuming your DataFrame is named 'df' and the column you want to clean is named 'Age'

# Remove non-digit characters and keep only numbers
Data['Age'] = Data['Age'].astype(str).str.replace(r'\D+', '', regex=True)

# Replace occurrences of more than two numbers by separating them with whitespace
def replace_with_space(value):
    numbers = re.findall(r'\d+', value)
    if len(numbers) >= 2:
        value = ' '.join(numbers)
    return value

Data['Age'] = Data['Age'].apply(replace_with_space)

# Display unique values of the 'Age' column
unique_values = Data['Age'].unique()
print(unique_values)
```

```python
# CLEAN COLUMN Location :

Local = Data['Location']

cleaned_location = []
for word in Local:
    if str(word) == 'nan' or str(word).strip() == '':
        cleaned_location.append('Unknown')
    else:
        cleaned_word = re.sub('[^a-zA-Z ]', '', str(word))  # Remove symbols
        cleaned_word = cleaned_word.title()  # Capitalize first letter of each word
        cleaned_location.append(cleaned_word)

data_shape = np.shape(cleaned_location)

# Count the unique values in the cleaned_name list
unique_value_count = len(set(cleaned_location))

print("Shape of cleaned_name:", data_shape)
print("Unique value count:", unique_value_count)
```
