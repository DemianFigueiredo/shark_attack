```python
# Get the list of column names in the DataFrame
columns = Data.columns

# Flag to indicate if any columns are equal to another column
are_columns_equal = False

# Iterate over each combination of columns
for i in range(len(columns)):
    for j in range(i+1, len(columns)):
        column1 = columns[i]
        column2 = columns[j]
        
        # Check if any value in column1 is equal to column2
        if any(Data[column1] == Data[column2]):
            print(f"At least one value in {column1} is equal to {column2}.")
            are_columns_equal = True

# If no columns are equal to each other
if not are_columns_equal:
     print("No columns are equal to each other.")
```

```python
found_column = False

# Iterate over each column
for column1 in Data.columns:
    for column2 in Data.columns:
        if column1 != column2:  # Skip self-comparison
            # Check if all values in column1 are equal to column2
            if (Data[column1] == Data[column2]).all():
                print(f"All values in column '{column1}' are equal to all values in column '{column2}'.")
                found_column = True

# If no column meets the condition
if not found_column:
    print("No column has all values equal to any other column.")
```

```python
# CLEAN COLUMN Origina Order:

Data['original order'] = Data['original order'].astype(str)

# Check for empty spaces in 'original order' column
has_empty_spaces = Data['original order'].str.contains(' ').any()

# Check for duplicated values in 'original order' column
has_duplicates = Data['original order'].duplicated().any()

# Check for NaN values in 'original order' column
has_nan_values = Data['original order'].isnull().any()

# Check if all values in 'original order' column are unique
all_unique_values = Data['original order'].nunique() == Data['original order'].shape[0]

print("Empty spaces in 'original order' column: ", has_empty_spaces)
print("Duplicated values in 'original order' column: ", has_duplicates)
print("NaN values in 'original order' column: ", has_nan_values)
print("All values in 'original order' column are unique: ", all_unique_values)


```

```python
# CLEAN COLUMN Date:

DataX['Date'] = pd.to_datetime(DataX['Date'], errors='coerce')

# Replace single whitespace with an empty string
DataX['Date'] = DataX['Date'].replace(' ', '')

# Transform day-month-year values into day-month(name) format
for i, date_val in enumerate(DataX['Date']):
    if pd.notnull(date_val) and date_val.day > 12:
        month_name = date_val.strftime('%B')  # Get month name
        DataX.loc[i, 'Date'] = date_val.strftime(f'%d-{month_name}')
    
# Replace NaT and NaN values with 'Undefined'
DataX['Date'].fillna('Undefined', inplace=True)



DataX.head()
```
